```python
# Input
x = bankrupt.iloc[:,:-1]

# Target variable

y = bankrupt.iloc[:,-1]
```


<p style= "background-color:#e9fafc;font-family:Georgia;color:#000000;font-size:100%;text-align:center;border-radius:10px 10px"><b>Sliptting the data into train and test.</b></p>

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.20, random_state = 0)
```


<p style= "background-color:#000000;font-family:Georgia;color:#FFFFFF;font-size:125%;text-align:center;border-radius:10px 10px;border-style:solid;border-width:3px;border-color:#000000;"><b>Preparing models</b></p>

<header><p style= "background-color:#e9fafc;font-family:Georgia;color:#000000;font-size:100%;text-align:center;border-radius:10px 10px"><b>1. Logistic Regression</b></p></header> 

```python
from sklearn.linear_model import LogisticRegression

logisticlassifier = LogisticRegression() 
```

```python
logisticlassifier.fit(x_train, y_train)

logisticlassifier.coef_ # coefficients of features
```

# After training the model we are trying model on test data

```python
y_pred = logisticlassifier.predict(x_test)
y_pred
```

# let's test the performance of our model - confusion matrix

```python
from sklearn.metrics import confusion_matrix

confusion_logist = confusion_matrix(y_test, y_pred)

confusion_logist
```

# As we can see No false positive & no false negative.

```python
#Accuracy of a Model
# Train Accuracy

train_acc_logist = np.mean(logisticlassifier.predict(x_train)== y_train)
train_acc_logist
```

```python
# Test Accuracy

test_acc_logist = np.mean(logisticlassifier.predict(x_test)== y_test)
test_acc_logist
```

```python
from sklearn.metrics import accuracy_score

logistic_acc = accuracy_score(y_test, y_pred)
logistic_acc
```

It turns out they have different but equally useful properties. From a practical standpoint, L1 tends to shrink coefficients to zero whereas L2 tends to shrink coefficients evenly. L1 is therefore useful for feature selection, as we can drop any variables associated with coefficients that go to zero.
